# Image classification on the MNIST dataset.

```python
import torchvision
from modules.model import train, eval, ConvolutionalModel
import torch
from torch.utils.data import DataLoader
```

Importing the dataset ,splitting it into the train, valid and test subsets and initializing the data loaders.

```python
mnist_dataset = torchvision.datasets.MNIST('/mnist_data/', download=True, transform=torchvision.transforms.ToTensor())

train_size=int(0.6*len(mnist_dataset))
valid_size=int(0.2*len(mnist_dataset))
test_size=len(mnist_dataset)-(train_size+valid_size)

mnist_train_dataset, mnist_valid_dataset, mnist_test_dataset = torch.utils.data.random_split(mnist_dataset, [train_size, valid_size, test_size])

train_dataloader = DataLoader(mnist_train_dataset, batch_size=50, shuffle=True, num_workers=2)
valid_dataloader = DataLoader(mnist_valid_dataset, batch_size=50, shuffle=True, num_workers=2)
test_dataloader = DataLoader(mnist_test_dataset, batch_size=50, num_workers=2)
```

Initializing the model, optimizer and criterion.

```python
model=ConvolutionalModel(1, 16, 32, 512, 10)

optimizer=torch.optim.SGD([
        {"params": [*model.conv1.parameters(),*model.conv2.parameters(),*model.fc1.parameters()], "weight_decay": 0.001},
        {"params": model.fc_logits.parameters(), "weight_decay": 0.}
    ], lr=0.001)

criterion = torch.nn.BCEWithLogitsLoss()
```

Training the model.

```python
for i in range(50):
    print(f'Epoch {i+1}:')
    train_loss = train(model, train_dataloader, criterion, optimizer)
    print(f'Training loss = {train_loss}')

    valid_loss, valid_acc = eval(model, valid_dataloader, criterion)
    print(f'Validation loss = {valid_loss}')
    print(f'Validation accuracy = {valid_acc}')
    print()
```

It is now time to test the model on the test set.

```python
loss, acc=eval(model, test_dataloader, criterion)

print(f"Test loss={loss}, test accuracy={acc}.")
```
